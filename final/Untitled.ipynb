```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import Data From CSV and Basic Analysis

```python
f_2005 = pd.read_csv("2005.csv")
f_2006 = pd.read_csv("2006.csv")
f_2007 = pd.read_csv("2007.csv")
f_2008 = pd.read_csv("2008.csv")
f_2009 = pd.read_csv("2009.csv")
```

Sale Number Trend

```python
List = [sum(f_2005["NUMBER OF SALES"]), sum(f_2006["NUMBER OF SALES"]), sum(f_2007["NUMBER OF SALES"]), sum(f_2008["NUMBER OF SALES"]), sum(f_2009["NUMBER OF SALES"])]
plt.plot(range(2005, 2010), List)
plt.show()
```

Sale Total Price Trend

```python
def value(var):
    return sum(var["AVERAGE SALE PRICE"]*var["NUMBER OF SALES"])
List = [value(f_2005), value(f_2006), value(f_2007), value(f_2008), value(f_2009)]
plt.plot(range(2005, 2010), List)
plt.show()
```

Sale Average Price Trend

```python
def value(var):
    return sum(var["AVERAGE SALE PRICE"]*var["NUMBER OF SALES"] / sum(var["NUMBER OF SALES"]))
List = [value(f_2005), value(f_2006), value(f_2007), value(f_2008), value(f_2009)]
plt.plot(range(2005, 2010), List)
plt.show()
```

Change All Text Format Data Into Num For Avg Price Prediction

```python
def style(var, year):
    var['YEAR'] = year
    var = var.join(pd.get_dummies(var["NEIGHBORHOOD"])).drop(["NEIGHBORHOOD"], axis=1)
    var = var.join(pd.get_dummies(var["TYPE OF HOME"])).drop(["TYPE OF HOME"], axis=1)
    var = var.drop(["NUMBER OF SALES", "LOWEST SALE PRICE", "MEDIAN SALE PRICE", "HIGHEST SALE PRICE"], axis=1)
    return var

f_2005 = style(f_2005, 2005)
f_2006 = style(f_2006, 2006)
f_2007 = style(f_2007, 2007)
f_2008 = style(f_2008, 2008)
f_2009 = style(f_2009, 2009)
```

```python
f_2005.head()
```

```python
total = pd.concat([f_2005, f_2006, f_2007, f_2008, f_2009]).drop(["TOTAL NO. OF PROPERTIES"], axis=1)
total.head()
```

```python
y = total["AVERAGE SALE PRICE"]
x = total.drop(["AVERAGE SALE PRICE"], axis=1)
x.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
model.score(x, y)
```
